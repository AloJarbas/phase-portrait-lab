## What is in each figure

Each generated SVG combines:

- a normalized vector field,
- the nullclines for $\dot{x}=0$ and $\dot{y}=0$,
- a few seeded trajectories,
- marked fixed points.

That is enough structure to ask useful first questions:

1. where does the flow slow down,
2. where do the nullclines intersect,
3. whether nearby trajectories are pulled in, pushed away, or wrapped around.

## Three systems, three kinds of behavior

### Van der Pol
The origin is not the whole story. The seeded trajectories show why people call this a self-excited oscillator: trajectories are pushed toward a closed orbit instead of simply collapsing to equilibrium.

### Lotka-Volterra
The coexistence point is surrounded by cycling trajectories. In this simplified model the orbits are closed, which makes it a good first picture and a bad stopping point if you care about dissipation or realism.

### Brusselator
This is the chemistry-flavored one. It is small enough to draw cleanly and still shows why a fixed point can stop being the only interesting structure in the plane.
