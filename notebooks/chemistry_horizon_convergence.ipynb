# Chemistry horizon convergence

This notebook slows down one narrow point from the chemistry lane in `phase-portrait-lab`: **the Hopf threshold can be exact while the measured newborn cycle size is still horizon-sensitive.**

That is the real reason the repo now keeps local stability and finite-time orbit measurement separate.


## 1. What the Jacobian gives and what it does not

For the Brusselator at fixed `A`, the local Hopf edge is

$$B = 1 + A^2.$$

For the Selkov model at fixed `a < 1/8`, the lower and upper Hopf edges form a finite oscillatory band.

Those thresholds are exact local facts. But a finite RK4 run still has to answer a separate question: how large does the visible cycle look once the orbit has had time to settle?


```python
from phaseportraitlab.chemistry_horizon_compare import build_brusselator_horizon_rows, build_selkov_horizon_rows

bruss = build_brusselator_horizon_rows()
selkov = build_selkov_horizon_rows()
bruss[0], selkov[0]
```

## 2. Near threshold, short horizons inflate amplitude

The new sidecar keeps the comparison deliberately unfair in one specific way: it uses a **short** horizon and a **long** horizon for the same parameters.

If the two answers disagree near onset, the disagreement is not coming from local theory. It is coming from finite-time measurement.


```python
[(row.edge_distance, round(row.short_x_amplitude, 3), round(row.long_x_amplitude, 3)) for row in bruss[:4]]
```

```python
[(row.edge_distance, round(row.short_x_amplitude, 3), round(row.long_x_amplitude, 3)) for row in selkov[:4]]
```

## 3. Period settles earlier than amplitude

The sharper part of the result is not just that the short horizon is wrong near onset.

It is that the **period estimate is already fairly stable** while the **amplitude estimate still is not**. That means the first visible cycle-size numbers near threshold deserve more skepticism than the first visible period numbers.


```python
max_period_gap = max(row.period_gap for row in bruss + selkov)
max_period_gap
```
