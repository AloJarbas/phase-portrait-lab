# Pendulum nonlinear saddle sidecar

This notebook is the slower companion to `reports/pendulum-nonlinear-saddle.md`.

The whole point is narrow: the upright pendulum is still a saddle, but its separatrix is an energy contour, not a pair of straight global eigenlines.


## 1. Load the sampled branch table

The CSV sidecar stores the positive branch `ω = 2 sin(φ/2)` beside the linear tangent `ω = φ`.


```python
import csv
from pathlib import Path

rows = []
with (Path('..') / 'assets' / 'pendulum-nonlinear-saddle.csv').open() as handle:
    for row in csv.DictReader(handle):
        rows.append({key: float(value) for key, value in row.items()})
rows[:3], rows[-3:]
```

## 2. Re-state the exact object

For the undamped pendulum, `H(θ, ω) = 1 - cos θ + ω²/2`. The upright equilibrium has `H = 2`, so the separatrix is the level set `H = 2`.


```python
from phaseportraitlab.nonlinear_saddle import pendulum_energy

samples = []
for row in rows[::40]:
    phi = row['shift']
    theta = 3.141592653589793 + phi
    omega = row['exact_omega']
    samples.append((round(phi, 4), round(pendulum_energy(theta, omega), 6)))
samples
```

## 3. Read the approximation gap

The local tangent is the right first derivative story, but the global branch keeps bending away from it.


```python
for row in rows[::45]:
    print(f"phi={row['shift']:.3f} exact={row['exact_omega']:.3f} linear={row['linear_omega']:.3f} error={row['abs_error']:.3f}")
```
