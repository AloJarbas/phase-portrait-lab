# Local linearization and the Brusselator Hopf threshold

This notebook is the slower companion to the gallery card in `assets/brusselator-hopf-sweep.svg`.

The point is narrow: for the Brusselator with fixed `A = 1`, the same fixed point changes character as `B` crosses `1 + A^2 = 2`.

That is the local story behind the larger phase portrait: the arrows still rotate, but the damping flips sign.

## Fixed point and Jacobian

For

$$\dot x = A + x^2 y - (B + 1)x, \qquad \dot y = Bx - x^2 y,$$

the fixed point is

$$ (x^*, y^*) = (A, B/A). $$

At that point the Jacobian is

$$ J(x^*, y^*) = \begin{bmatrix} B - 1 & A^2 \\ -B & -A^2 \end{bmatrix}. $$

So the trace and determinant are

$$ \operatorname{tr}(J) = B - 1 - A^2, \qquad \det(J) = A^2. $$

With `A = 1`, the determinant stays positive and the sign change comes entirely from the trace: `tr(J) = B - 2`.

```python
from phaseportraitlab.brusselator_sweep import brusselator_hopf_threshold, sweep_brusselator_b_values

threshold = brusselator_hopf_threshold(1.0)
rows = sweep_brusselator_b_values([1.6, threshold, 2.4])
[(row.b, row.trace, row.classification) for row in rows]
```

## Three anchor settings

- **`B = 1.6`**: negative trace, so the fixed point is a stable spiral
- **`B = 2.0`**: zero trace, so the linearization lands on the Hopf threshold
- **`B = 2.4`**: positive trace, so the fixed point becomes an unstable spiral and the large orbit wins

That is why the same portrait family can move from calm local damping to self-sustained oscillation without changing the basic reaction form.

```python
for row in rows:
    eig = row.eigenvalues[0]
    print(
        f"B={row.b:.2f}  trace={row.trace:+.2f}  Re(lambda)={eig.real:+.2f}  |Im(lambda)|={abs(eig.imag):.2f}  {row.classification}"
    )
```

## What the sweep figure adds

The gallery card places three readings on one canvas:

1. the Jacobian trace crossing zero
2. the real part of the eigenvalues switching from damping to growth
3. the oscillation frequency staying present while the stability changes sign

That is the key teaching move here: **oscillation and stability are not the same question**.

A complex pair can keep the turning motion visible on both sides of the threshold while only the real part decides whether nearby trajectories spiral in or out.
